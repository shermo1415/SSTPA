```python
import numpy as np
import matplotlib.pyplot as plt
from config import configs
```

```python
pred = np.load('result/pred.npy')
true = np.load('result/true.npy')
i1i2 = np.load('result/i1i2.npy')
i1 = i1i2[:,:len(configs.dims)]
i2 = i1i2[:,len(configs.dims):]
diff = np.abs(pred - true)
```

```python
i1.shape, i2.shape
```

```python
#single frame
sam = 80
dim = 2

plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.imshow(i1[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.subplot(132)
plt.imshow(pred[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.subplot(133)
plt.imshow(i2[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()

plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.imshow(i1[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.subplot(132)
plt.imshow(true[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.subplot(133)
plt.imshow(i2[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()

plt.figure(figsize=(12, 4))
plt.imshow(diff[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
diff.mean()
```

```python
#multi frames
sam = 130
dim = 2
frames = 3

print('pred')
plt.figure(figsize=(12, 4))
plt.subplot(100+(frames+2)*10+1)
plt.imshow(i1[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
for f in range(frames):
    plt.subplot(100+(frames+2)*10+f+2)
    plt.imshow(pred[sam, frames * f + dim], cmap = 'jet')
    plt.xticks([])
    plt.yticks([])
    plt.axis('off')
plt.subplot(100+(frames+2)*10+frames+2)
plt.imshow(i2[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()

print('true')
plt.figure(figsize=(12, 4))
plt.subplot(100+(frames+2)*10+1)
plt.imshow(i1[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
for f in range(frames):
    plt.subplot(100+(frames+2)*10+f+2)
    plt.imshow(true[sam, frames * f + dim], cmap = 'jet')
    plt.xticks([])
    plt.yticks([])
    plt.axis('off')
plt.subplot(100+(frames+2)*10+frames+2)
plt.imshow(i2[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()

print('diff')
plt.figure(figsize=(12, 4))
plt.subplot(100+(frames+2)*10+1)
plt.imshow(i1[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
for f in range(frames):
    plt.subplot(100+(frames+2)*10+f+2)
    plt.imshow(diff[sam, frames * f + dim], cmap = 'jet')
    plt.xticks([])
    plt.yticks([])
    plt.axis('off')
plt.subplot(100+(frames+2)*10+frames+2)
plt.imshow(i2[sam, dim], cmap = 'jet')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```
